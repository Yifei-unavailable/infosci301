```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define GitHub raw URLs for the datasets
urls = {
    "avg_exposure": "https://raw.githubusercontent.com/Yifei-unavailable/infosci301/refs/heads/main/Research%20Proposal/Fig2a-avg_exposure.csv",
    "number_of_links": "https://raw.githubusercontent.com/Yifei-unavailable/infosci301/refs/heads/main/Research%20Proposal/Fig2b-number_of_links.csv",
    "distress_incomegroup": "https://raw.githubusercontent.com/Yifei-unavailable/infosci301/refs/heads/main/Research%20Proposal/Fig3a-distress_incomegroup.csv",
    "Ed_vs_GDPpc": "https://raw.githubusercontent.com/Yifei-unavailable/infosci301/refs/heads/main/Research%20Proposal/Fig3b-EdvsGDPpc.csv",
}

# Load datasets
avg_exposure = pd.read_csv(urls["avg_exposure"])
number_of_links = pd.read_csv(urls["number_of_links"])
distress_incomegroup = pd.read_csv(urls["distress_incomegroup"])
ed_vs_gdppc = pd.read_csv(urls["Ed_vs_GDPpc"])

# Quick data preview
print("Average Exposure Dataset:\n", avg_exposure.head())
print("\nNumber of Links Dataset:\n", number_of_links.head())
print("\nDistress by Income Group Dataset:\n", distress_incomegroup.head())
print("\nEducation vs GDP per Capita Dataset:\n", ed_vs_gdppc.head())

# Check column names
print("Column names for avg_exposure:", avg_exposure.columns)
print("Column names for number_of_links:", number_of_links.columns)
print("Column names for distress_incomegroup:", distress_incomegroup.columns)
print("Column names for ed_vs_gdppc:", ed_vs_gdppc.columns)

# Set Seaborn theme for consistent styling
sns.set_theme(style="whitegrid")

# 1. Redesign for Fig2a (Eigenvector Centrality - Buyers) and Fig2b (Sellers)
fig, ax = plt.subplots(1, 2, figsize=(18, 6), sharey=True)

# Buyers Plot
sns.barplot(
    data=avg_exposure,
    x=avg_exposure.columns[0],  # Dynamically use first column (e.g., Country)
    y=avg_exposure.columns[1],  # Dynamically use second column (e.g., Eigenvector Centrality)
    ax=ax[0],
    palette="viridis"
)
ax[0].set_title("Eigenvector Centrality (Buyers)", fontsize=14)
ax[0].set_xlabel(avg_exposure.columns[0], fontsize=12)
ax[0].set_ylabel(avg_exposure.columns[1], fontsize=12)
ax[0].tick_params(axis="x", rotation=90)

# Sellers Plot
sns.barplot(
    data=number_of_links,
    x=number_of_links.columns[0],  # Dynamically use first column
    y=number_of_links.columns[1],  # Dynamically use second column
    ax=ax[1],
    palette="viridis"
)
ax[1].set_title("Eigenvector Centrality (Sellers)", fontsize=14)
ax[1].set_xlabel(number_of_links.columns[0], fontsize=12)
ax[1].set_ylabel("", fontsize=12)  # Shared y-axis

plt.tight_layout()
plt.show()

# 2. Redesign for Fig3a (Distress by Income Group)
plt.figure(figsize=(10, 6))
sns.barplot(
    data=distress_incomegroup,
    x=distress_incomegroup.columns[0],  # Dynamically use first column (e.g., Income Group)
    y=distress_incomegroup.columns[1],  # Dynamically use second column (e.g., Distress Index)
    palette="coolwarm"
)
plt.title("Distress Index by Income Group", fontsize=16)
plt.xlabel(distress_incomegroup.columns[0], fontsize=12)
plt.ylabel(distress_incomegroup.columns[1], fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Redesign for Fig3b (Education vs GDP per Capita)
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=ed_vs_gdppc,
    x=ed_vs_gdppc.columns[0],  # Dynamically use first column (e.g., GDP per Capita)
    y=ed_vs_gdppc.columns[1],  # Dynamically use second column (e.g., Education Index)
    size=ed_vs_gdppc.columns[2],  # Dynamically use third column (e.g., Population)
    hue=ed_vs_gdppc.columns[3],  # Dynamically use fourth column (e.g., Region)
    palette="muted",
    sizes=(50, 500),
    alpha=0.7,
)
plt.title("Education Index vs GDP per Capita", fontsize=16)
plt.xlabel(ed_vs_gdppc.columns[0], fontsize=12)
plt.ylabel(ed_vs_gdppc.columns[1], fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left", title=ed_vs_gdppc.columns[3])
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Simulate data for Buyers and Sellers Eigenvector Centrality
years = np.arange(2000, 2015)
countries = ["China", "United States", "Japan", "Korea Republic", "Ireland"]
highlighted_countries = ["China", "United States", "Japan"]

# Simulated data for Buyers
buyers_data = {
    "Year": np.tile(years, len(countries)),
    "Country": np.repeat(countries, len(years)),
    "Eigenvector Centrality": np.concatenate([
        np.logspace(-3, 0, len(years)),   # China
        np.logspace(0, -2, len(years)),  # United States
        np.logspace(-1, -3, len(years)), # Japan
        np.logspace(-2, -4, len(years)), # Korea Republic
        np.logspace(-3, -5, len(years)), # Ireland
    ])
}
buyers_df = pd.DataFrame(buyers_data)

# Simulated data for Sellers
sellers_data = {
    "Year": np.tile(years, len(countries)),
    "Country": np.repeat(countries, len(years)),
    "Eigenvector Centrality": np.concatenate([
        np.logspace(-2, 0, len(years)),  # China
        np.logspace(0, -1, len(years)), # United States
        np.logspace(-1, -2, len(years)),# Japan
        np.logspace(-2, -3, len(years)),# Korea Republic
        np.logspace(-3, -4, len(years)),# Ireland
    ])
}
sellers_df = pd.DataFrame(sellers_data)

# Plot improved visualizations
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Highlighted colors for key countries
highlight_colors = {
    "China": "red",
    "United States": "blue",
    "Japan": "orange"
}
default_color = "gray"

# Plot for Buyers
for country in countries:
    subset = buyers_df[buyers_df["Country"] == country]
    color = highlight_colors[country] if country in highlight_colors else default_color
    lw = 2.5 if country in highlighted_countries else 1
    zorder = 3 if country in highlighted_countries else 1
    axes[0].plot(
        subset["Year"], subset["Eigenvector Centrality"],
        label=country if country in highlighted_countries else None,
        color=color, linewidth=lw, zorder=zorder, alpha=0.8
    )

axes[0].set_title("Buyers (Improved Visualization)", fontsize=16)
axes[0].set_xlabel("Year", fontsize=12)
axes[0].set_ylabel("Eigenvector Centrality", fontsize=12)
axes[0].set_yscale("log")
axes[0].legend(title="Highlighted Countries", fontsize=10, loc="upper left")
axes[0].axvline(2008, color="black", linestyle="--", linewidth=1, alpha=0.7)
axes[0].text(2008, 10**-0.5, "2008 Financial Crisis", rotation=90, color="black", fontsize=10)

# Plot for Sellers
for country in countries:
    subset = sellers_df[sellers_df["Country"] == country]
    color = highlight_colors[country] if country in highlight_colors else default_color
    lw = 2.5 if country in highlighted_countries else 1
    zorder = 3 if country in highlighted_countries else 1
    axes[1].plot(
        subset["Year"], subset["Eigenvector Centrality"],
        label=country if country in highlighted_countries else None,
        color=color, linewidth=lw, zorder=zorder, alpha=0.8
    )

axes[1].set_title("Sellers (Improved Visualization)", fontsize=16)
axes[1].set_xlabel("Year", fontsize=12)
axes[1].set_yscale("log")
axes[1].legend(title="Highlighted Countries", fontsize=10, loc="upper left")
axes[1].axvline(2008, color="black", linestyle="--", linewidth=1, alpha=0.7)
axes[1].text(2008, 10**-0.5, "2008 Financial Crisis", rotation=90, color="black", fontsize=10)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# Simulate data for Buyers and Sellers Eigenvector Centrality
years = np.arange(2000, 2015)
countries = ["China", "United States", "Japan", "Korea Republic", "Ireland"]
highlighted_countries = ["China", "United States", "Japan"]

# Simulated data for Buyers
buyers_data = {
    "Year": np.tile(years, len(countries)),
    "Country": np.repeat(countries, len(years)),
    "Eigenvector Centrality": np.concatenate([
        np.logspace(-3, 0, len(years)),   # China
        np.logspace(0, -2, len(years)),  # United States
        np.logspace(-1, -3, len(years)), # Japan
        np.logspace(-2, -4, len(years)), # Korea Republic
        np.logspace(-3, -5, len(years)), # Ireland
    ])
}
buyers_df = pd.DataFrame(buyers_data)

# Simulated data for Sellers
sellers_data = {
    "Year": np.tile(years, len(countries)),
    "Country": np.repeat(countries, len(years)),
    "Eigenvector Centrality": np.concatenate([
        np.logspace(-2, 0, len(years)),  # China
        np.logspace(0, -1, len(years)), # United States
        np.logspace(-1, -2, len(years)),# Japan
        np.logspace(-2, -3, len(years)),# Korea Republic
        np.logspace(-3, -4, len(years)),# Ireland
    ])
}
sellers_df = pd.DataFrame(sellers_data)

# Add a column for Highlight Status
buyers_df["Highlight"] = buyers_df["Country"].apply(lambda x: "Highlighted" if x in highlighted_countries else "Other")
sellers_df["Highlight"] = sellers_df["Country"].apply(lambda x: "Highlighted" if x in highlighted_countries else "Other")

# Interactive Plot for Buyers
fig_buyers = px.line(
    buyers_df,
    x="Year",
    y="Eigenvector Centrality",
    color="Country",
    line_group="Country",
    line_dash="Highlight",
    title="Advanced Buyers Visualization (Interactive)",
    log_y=True,  # Logarithmic scale
    template="plotly_white",
    labels={"Eigenvector Centrality": "Eigenvector Centrality (log scale)"}
)

# Add an annotation for the 2008 financial crisis
fig_buyers.add_vline(
    x=2008,
    line_width=2,
    line_dash="dash",
    line_color="black",
    annotation_text="2008 Financial Crisis",
    annotation_position="top right"
)

# Interactive Plot for Sellers
fig_sellers = px.line(
    sellers_df,
    x="Year",
    y="Eigenvector Centrality",
    color="Country",
    line_group="Country",
    line_dash="Highlight",
    title="Advanced Sellers Visualization (Interactive)",
    log_y=True,  # Logarithmic scale
    template="plotly_white",
    labels={"Eigenvector Centrality": "Eigenvector Centrality (log scale)"}
)

# Add an annotation for the 2008 financial crisis
fig_sellers.add_vline(
    x=2008,
    line_width=2,
    line_dash="dash",
    line_color="black",
    annotation_text="2008 Financial Crisis",
    annotation_position="top right"
)

# Show the plots
fig_buyers.show()
fig_sellers.show()
```

```python
import plotly.express as px
import numpy as np
import pandas as pd

# Simulate data
years = np.arange(2000, 2015)
countries = ["China", "United States", "Japan", "Korea Republic", "Ireland"]

data = {
    "Year": np.tile(years, len(countries)),
    "Country": np.repeat(countries, len(years)),
    "Eigenvector Centrality": np.concatenate([
        np.logspace(-3, 0, len(years)),  # China
        np.logspace(0, -2, len(years)),  # US
        np.logspace(-1, -3, len(years)), # Japan
        np.logspace(-2, -4, len(years)), # Korea
        np.logspace(-3, -5, len(years))  # Ireland
    ]),
}

df = pd.DataFrame(data)

# Plot Buyers' Centrality
fig = px.line(
    df,
    x="Year",
    y="Eigenvector Centrality",
    color="Country",
    log_y=True,
    title="Interactive Visualization: Buyers' Eigenvector Centrality",
    template="plotly_white",
    labels={"Eigenvector Centrality": "Eigenvector Centrality (log scale)"}
)

# Add annotation for 2008 financial crisis
fig.add_vline(x=2008, line_width=2, line_dash="dash", line_color="black")
fig.add_annotation(
    x=2008,
    y=0.1,
    text="2008 Financial Crisis",
    showarrow=True,
    arrowhead=1
)

fig.show()
```
